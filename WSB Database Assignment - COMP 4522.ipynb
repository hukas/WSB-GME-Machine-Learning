```python
import pandas as pd
from pandas import DataFrame
import altair as alt
import spacy as sp
# python -m spacy download en_core_web_sm
sp.load('en_core_web_sm')
wsb = pd.read_csv("reddit_wsb.csv")
gme = pd.read_csv("GMEday.csv")
gme.tail(2)
```

```python
wsb["timestamp"] = pd.to_datetime(wsb["timestamp"], errors='coerce')
gme["Date"] = pd.to_datetime(gme["Date"], errors='coerce')
#wsb["timestamp"].dt.day
#wsb["timestamp"].dt.month
#wsb["timestamp"].dt.year

wsbSimple = wsb.loc[:, ['id', 'title', 'score', 'timestamp']]
wsbSimple = wsbSimple.assign(timestamp=wsbSimple.timestamp.dt.round('D'))
wsbGME = pd.merge(left=wsbSimple, right=gme, how='left', left_on='timestamp', 
                  right_on='Date')

wsbGME = wsbGME.sort_values(by='timestamp', ascending=True)
wsbGME = wsbGME.fillna(method='ffill')
wsbGME = wsbGME.rename(columns={'timestamp' : 'PostDate', 'title' : 'PostTitle', 
                                'score' : 'PostScore', 'Date' : 'MarketDate', 'id' : 'ID'})
#wsbGME.isnull().any()
wsbGME
```
